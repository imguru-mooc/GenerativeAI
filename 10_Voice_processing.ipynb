###FFT 직접 구현 (Discrete Fourier Transform, DFT 방식)

```python
# 필요한 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt

# 간단한 DFT (Discrete Fourier Transform) 직접 구현
def simple_fft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)  # 복소수 결과 배열 준비

    for k in range(N):  # 각 주파수 성분 k에 대해
        for n in range(N):  # 시간 성분 n에 대해 누적 합산
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)

    return X

# 샘플 데이터 생성
HZ=256
t = np.linspace(0, 1, HZ, endpoint=False)  # 1초 동안 64개 샘플
signal_5hz = np.sin(2 * np.pi * 5 * t)     # 5Hz 사인파
signal_10hz = 0.5 * np.sin(2 * np.pi * 10 * t)  # 10Hz 사인파 (진폭 0.5)

# 합성 신호 (5Hz + 10Hz)
signal_combined = signal_5hz + signal_10hz

# 직접 만든 FFT 적용
fft_result = simple_fft(signal_combined)

# 주파수 축 생성
sample_rate = HZ  # 64샘플/초
freqs = np.fft.fftfreq(len(t), 1/sample_rate)

# 그래프 그리기
plt.figure(figsize=(14, 10))

# 1. 5Hz 사인파
plt.subplot(4, 1, 1)
plt.plot(t, signal_5hz)
plt.title('5 Hz Sine Wave (Time Domain)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

# 2. 10Hz 사인파
plt.subplot(4, 1, 2)
plt.plot(t, signal_10hz, color='orange')
plt.title('10 Hz Sine Wave (Time Domain)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

# 3. 합성 신호 (5Hz + 10Hz)
plt.subplot(4, 1, 3)
plt.plot(t, signal_combined, color='green')
plt.title('Combined Signal (5 Hz + 10 Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

# 4. 합성 신호의 FFT 결과
plt.subplot(4, 1, 4)
plt.stem(freqs[:len(freqs)//2], np.abs(fft_result)[:len(freqs)//2])
plt.title('FFT Spectrum of Combined Signal (Frequency Domain)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

###STFT (Short-Time Fourier Transform)함수 구현

```python
# 필요한 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import get_window

# STFT 함수 직접 정의
def compute_stft(x, fs, window='hann', nperseg=512, noverlap=256):
    """
    x : 입력 신호
    fs : 샘플링 주파수
    window : 윈도우 종류 (예: 'hann')
    nperseg : 한 프레임 길이
    noverlap : 프레임 간 중첩 샘플 수
    """
    hop_size = nperseg - noverlap
    window_func = get_window(window, nperseg)
    num_frames = (len(x) - noverlap) // hop_size

    stft_matrix = []

    for i in range(num_frames):
        start = i * hop_size
        end = start + nperseg
        frame = x[start:end]
        if len(frame) < nperseg:
            frame = np.pad(frame, (0, nperseg - len(frame)))
        windowed_frame = frame * window_func
        spectrum = np.fft.fft(windowed_frame)
        stft_matrix.append(spectrum)

    stft_matrix = np.array(stft_matrix).T  # 주파수 x 시간 구조로
    time_axis = np.arange(num_frames) * hop_size / fs
    freq_axis = np.fft.fftfreq(nperseg, 1/fs)

    return freq_axis[:nperseg//2], time_axis, np.abs(stft_matrix[:nperseg//2, :])

# 샘플 데이터 생성
t = np.linspace(0, 2, 4096, endpoint=False)  # 2초 동안 4096 포인트
fs = 2048  # 샘플링 주파수 (Hz)

# 100Hz + 200Hz 합성 신호
signal_100hz = np.sin(2 * np.pi * 100 * t)
signal_200hz = 0.5 * np.sin(2 * np.pi * 200 * t)
signal = signal_100hz + signal_200hz

# STFT 계산
freqs, times, magnitude = compute_stft(signal, fs, window='hann', nperseg=512, noverlap=256)

# 시각화
plt.figure(figsize=(14, 6))
plt.pcolormesh(times, freqs, 20*np.log10(magnitude + 1e-10), shading='gouraud', cmap='plasma')
plt.title('STFT Spectrogram (100Hz + 200Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Frequency [Hz]')
plt.colorbar(label='Intensity [dB]')
plt.ylim(0, 400)  # 주파수 범위 0~400Hz로 확장
plt.tight_layout()
plt.show()
```

### 멜 스펙토그램 변환 후 시각화

```python
# 필요한 라이브러리
import numpy as np
import matplotlib.pyplot as plt

# 1. Mel-Scale 변환 함수
def hz_to_mel(hz):
    return 2595 * np.log10(1 + hz / 700)

def mel_to_hz(mel):
    return 700 * (10**(mel / 2595) - 1)

# 2. Mel 필터 뱅크 직접 생성
def create_mel_filterbank(n_filters=128, n_fft=512, fs=2048, fmin=0, fmax=None):
    if fmax is None:
        fmax = fs / 2

    # Mel-scale 기준 포인트 생성
    mel_min = hz_to_mel(fmin)
    mel_max = hz_to_mel(fmax)
    mel_points = np.linspace(mel_min, mel_max, n_filters + 2)
    hz_points = mel_to_hz(mel_points)

    bin_freqs = np.fft.rfftfreq(n_fft, d=1/fs)
    bin_numbers = np.floor((n_fft + 1) * hz_points / fs).astype(int)

    filterbank = np.zeros((n_filters, len(bin_freqs)))

    for i in range(1, n_filters + 1):
        f_left = bin_numbers[i - 1]
        f_center = bin_numbers[i]
        f_right = bin_numbers[i + 1]

        if f_center > f_left:
            filterbank[i-1, f_left:f_center] = (bin_freqs[f_left:f_center] - hz_points[i-1]) / (hz_points[i] - hz_points[i-1])
        if f_right > f_center:
            filterbank[i-1, f_center:f_right] = (hz_points[i+1] - bin_freqs[f_center:f_right]) / (hz_points[i+1] - hz_points[i])

    return filterbank

# 3. STFT 함수 (rfft 버전)
def compute_stft_rfft(x, fs, nperseg=512, noverlap=256):
    hop_size = nperseg - noverlap
    num_frames = (len(x) - noverlap) // hop_size
    stft_matrix = []

    for i in range(num_frames):
        start = i * hop_size
        end = start + nperseg
        frame = x[start:end]
        if len(frame) < nperseg:
            frame = np.pad(frame, (0, nperseg - len(frame)))
        windowed_frame = frame * np.hanning(nperseg)
        spectrum = np.fft.rfft(windowed_frame)
        stft_matrix.append(spectrum)

    stft_matrix = np.array(stft_matrix).T  # (freq bins, time frames)
    time_axis = np.arange(num_frames) * hop_size / fs
    freq_axis = np.fft.rfftfreq(nperseg, 1/fs)

    return freq_axis, time_axis, np.abs(stft_matrix)

# 4. 신호 생성
t = np.linspace(0, 2, 4096, endpoint=False)
fs = 2048  # 샘플링 주파수
signal = np.sin(2 * np.pi * 100 * t) + 0.5 * np.sin(2 * np.pi * 200 * t)

# 5. STFT 계산
freqs, times, magnitude = compute_stft_rfft(signal, fs, nperseg=512, noverlap=256)

# 6. Mel 필터 뱅크 생성
mel_filterbank = create_mel_filterbank(n_filters=128, n_fft=512, fs=fs)

# 7. Mel 스펙트로그램 계산
mel_spectrogram = np.dot(mel_filterbank, magnitude)

# 8. dB 스케일 변환
mel_spectrogram_db = 20 * np.log10(mel_spectrogram + 1e-10)

# 9. 시각화
plt.figure(figsize=(14, 6))
plt.imshow(mel_spectrogram_db,
           aspect='auto',
           origin='lower',
           extent=[times.min(), times.max(), 0, 128],
           cmap='plasma')
plt.title('Mel Spectrogram (100Hz + 200Hz) [Direct Implementation]')
plt.xlabel('Time [s]')
plt.ylabel('Mel Filter Bank Index')
plt.colorbar(label='Intensity [dB]')
plt.tight_layout()
plt.show()
```

```python
# 필요한 라이브러리
import numpy as np
import matplotlib.pyplot as plt

# 1. Hz ↔ Mel 변환 함수
def hz_to_mel(hz):
    return 2595 * np.log10(1 + hz / 700)

def mel_to_hz(mel):
    return 700 * (10**(mel / 2595) - 1)

# 2. Mel 필터 뱅크 생성 함수
def create_mel_filterbank(n_filters=20, n_fft=512, fs=16000, fmin=0, fmax=None):
    if fmax is None:
        fmax = fs / 2

    # Mel 포인트 계산
    mel_min = hz_to_mel(fmin)
    mel_max = hz_to_mel(fmax)
    mel_points = np.linspace(mel_min, mel_max, n_filters + 2)
    hz_points = mel_to_hz(mel_points)

    # FFT bin에 대응되는 인덱스 계산
    bin_freqs = np.fft.rfftfreq(n_fft, 1/fs)
    bin_indices = np.floor((n_fft + 1) * hz_points / fs).astype(int)

    filterbank = np.zeros((n_filters, len(bin_freqs)))

    for i in range(1, n_filters + 1):
        left = bin_indices[i-1]
        center = bin_indices[i]
        right = bin_indices[i+1]

        if center > left:
            filterbank[i-1, left:center] = (bin_freqs[left:center] - hz_points[i-1]) / (hz_points[i] - hz_points[i-1])
        if right > center:
            filterbank[i-1, center:right] = (hz_points[i+1] - bin_freqs[center:right]) / (hz_points[i+1] - hz_points[i])

    return bin_freqs, filterbank

# 3. 샘플 파라미터 설정
fs = 16000        # 샘플링 주파수 (16kHz, 음성 기준)
n_fft = 512       # FFT 포인트 수
n_mels = 20       # Mel 필터 수 (예: 20개)
fmin = 0          # 최소 주파수
fmax = fs // 2    # 최대 주파수 (Nyquist)

# 4. 필터 뱅크 생성
bin_freqs, mel_filters = create_mel_filterbank(n_filters=n_mels, n_fft=n_fft, fs=fs, fmin=fmin, fmax=fmax)

# 5. 필터 뱅크 시각화
plt.figure(figsize=(12, 6))
for i in range(n_mels):
    plt.plot(bin_freqs, mel_filters[i])
plt.title('Mel Filter Bank (Triangular Filters)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.xlim(0, fs/2)
plt.tight_layout()
plt.show()
```

###MFCC 변환 후 시각화

```python
# 필요한 라이브러리
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft
from scipy.fftpack import dct

# 1. Hz ↔ Mel 변환 함수
def hz_to_mel(hz):
    return 2595 * np.log10(1 + hz / 700)

def mel_to_hz(mel):
    return 700 * (10**(mel / 2595) - 1)

# 2. Mel 필터 뱅크 생성 함수
def create_mel_filterbank(n_filters=20, n_fft=512, fs=16000, fmin=0, fmax=None):
    if fmax is None:
        fmax = fs / 2

    mel_min = hz_to_mel(fmin)
    mel_max = hz_to_mel(fmax)
    mel_points = np.linspace(mel_min, mel_max, n_filters + 2)
    hz_points = mel_to_hz(mel_points)

    bin_freqs = np.fft.rfftfreq(n_fft, 1/fs)
    bin_indices = np.floor((n_fft + 1) * hz_points / fs).astype(int)

    filterbank = np.zeros((n_filters, len(bin_freqs)))

    for i in range(1, n_filters + 1):
        left = bin_indices[i-1]
        center = bin_indices[i]
        right = bin_indices[i+1]
        if center > left:
            filterbank[i-1, left:center] = (bin_freqs[left:center] - hz_points[i-1]) / (hz_points[i] - hz_points[i-1])
        if right > center:
            filterbank[i-1, center:right] = (hz_points[i+1] - bin_freqs[center:right]) / (hz_points[i+1] - hz_points[i])

    return filterbank

# 3. 신호 생성
fs = 16000  # 샘플링 주파수
duration = 2.0  # 2초
t = np.linspace(0, duration, int(fs * duration), endpoint=False)
signal = np.sin(2 * np.pi * 440 * t)  # 440Hz (A4 음)

# 4. STFT (프레임 분할 + rFFT)
frame_size = 512
hop_size = 256
num_frames = (len(signal) - frame_size) // hop_size + 1

frames = np.zeros((num_frames, frame_size))
for i in range(num_frames):
    start = i * hop_size
    frames[i] = signal[start:start+frame_size] * np.hanning(frame_size)  # Hanning 윈도우 적용

# 5. Magnitude Spectrum
magnitude_spectrogram = np.abs(np.fft.rfft(frames, axis=1))

# 6. Mel 필터 뱅크 적용
mel_filters = create_mel_filterbank(n_filters=20, n_fft=frame_size, fs=fs)
mel_spectrogram = np.dot(magnitude_spectrogram, mel_filters.T)

# 7. Log 변환
log_mel_spectrogram = np.log(mel_spectrogram + 1e-10)

# 8. DCT 적용 (MFCC)
mfcc = dct(log_mel_spectrogram, type=2, axis=1, norm='ortho')

# 9. MFCC 시각화
plt.figure(figsize=(12, 6))
plt.imshow(mfcc.T,
           aspect='auto',
           origin='lower',
           extent=[0, duration, 0, 20],
           cmap='plasma')
plt.title('MFCC (from 440Hz Sine Wave)')
plt.xlabel('Time [s]')
plt.ylabel('MFCC Coefficients')
plt.colorbar(label='Coefficient Value')
plt.tight_layout()
plt.show()
```
