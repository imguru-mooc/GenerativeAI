##💼 기업 내 AI 적용 전략 실습

이 노트북은 기업 내부에서 생성형 AI를 어떻게 전략적으로 활용할 수 있는지 실습하고,
GPT 모델을 통해 실제 전략 요약, 업무 아이디어 도출, 실행계획 생성을 실험합니다.

###✏️ Step 1: 기업 개요 및 AI 활용 분야 입력

```python
company_info = """
이 회사는 중견 자동차 부품 제조업체로, 주요 제품은 전장 부품, 배터리 모듈 하우징, 차량 내 인포테인먼트 시스템이다.
최근 완성차 OEM의 디지털 수요 증가에 따라 생성형 AI 기술 도입을 검토 중이다.
관심 있는 분야는 기술 문서 자동화, 디자인 시각화, 품질 관리 보고서 생성, 고객 문의 대응 자동화이다.
"""
```

###🤖 Step 2: GPT를 활용한 AI 전략 아이디어 도출

```python
response = client.chat.completions.create(
  model="gpt-4.1-mini",
  messages=[
    {"role": "system", "content": "너는 자동차 산업 전문 AI 전략 컨설턴트야."},
    {"role": "user", "content": f"다음 자동차 부품 제조 기업에 대해 생성형 AI를 활용한 전략을 제안해줘:\n{company_info}"}
  ],
  temperature=0.7
)

print(response.choices[0].message.content)
```

###📊 Step 3: 부서별 AI 적용 아이디어 요약

```python
department_prompt = """
자동차 부품 제조업체의 주요 부서는 R&D, 생산, 품질관리, 고객지원, 마케팅이다.
각 부서별로 생성형 AI를 적용할 수 있는 실질적인 아이디어를 요약해줘.
"""

response2 = client.chat.completions.create(
  model="gpt-4.1-mini",
  messages=[
    {"role": "system", "content": "생성형 AI 전략 컨설턴트"},
    {"role": "user", "content": department_prompt}
  ]
)

print(response2.choices[0].message.content)
```

###📝 Step 4: 실행 계획 작성 템플릿 생성

```python
plan_prompt = """
위에서 제안된 전략과 부서별 아이디어를 기반으로, 자동차 부품 기업이 6개월 내 도입할 수 있는 생성형 AI 실행 로드맵을 작성해줘.
단계별 목표, 담당 부서, 활용 기술을 표로 정리해줘.
"""

response3 = client.chat.completions.create(
  model="gpt-4.1-mini",
  messages=[
    {"role": "user", "content": plan_prompt}
  ]
)

print(response3.choices[0].message.content)
```
